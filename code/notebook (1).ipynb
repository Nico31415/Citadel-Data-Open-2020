```
def eligible(year):
    popul = census_data[f'population_{year}'] >= 500
    income = census_data[f'medianHouseIncome_{year}'] < census_data[f'medianHouseIncome_{year}'].quantile(0.5)
    homeValue = census_data[f'medianHomeValue_{year}'] < census_data[f'medianHomeValue_{year}'].quantile(0.5)
    return popul & income & homeValue

for year in range(2009, 2019):
    census_data[f'isEligible_{year}'] = eligible(year)
```

```
for year in range(2009, 2019):
  print(census_data[f'isEligible_{year}'].value_counts())
```

```
def fractionEducated(year):
    bachelors = census_data[f'maleBachelors_{year}'] + census_data[f'femaleBachelors_{year}']
    masters = census_data[f'maleMasters_{year}'] + census_data[f'femaleMasters_{year}']
    doctorates = census_data[f'maleDoctoral_{year}'] + census_data[f'femaleDoctoral_{year}']
    return (bachelors + masters + doctorates) / (
            census_data[f'maleOver25_{year}'] + census_data[f'femaleOver25_{year}'])

for year in range (2009, 2019):
    census_data[f'fractionEducated_{year}'] = fractionEducated(year)
```

```
# adjust house prices for inflation according to 
# https://www.usinflationcalculator.com
def adjustHousePrices():
    adjustments = [1.00, 1.02, 1.05, 1.07, 1.09, 1.10, 1.10, 
                   1.12, 1.14, 1.17]
    for year in range(2009, 2019):
        census_data[f'adjustedHomeValue_{year}'] = census_data[f'medianHomeValue_{year}']/adjustments[year-2009]

adjustHousePrices()
```

```


#Test 2: Has the tract gentrified?
#- Educational Attainment (has to be in top 1/3 of the area): 
#  % of residents age >= 25 holding bachelors degrees
#- Tracts median home value increased when adjusted for inflation
#- % increase in tracts inflation adjusted median home was in the 
#. top 1/3 of tracts in that area
# testing every 5 years

# def refinedHasGentrified(census_data, years_apart, year):
#     census_data[f'homeValuePercentChange_{year}'] = census_data[f'medianHomeValue_{year}'] / census_data[f'medianHomeValue_{year - years_apart}'] - 1
#     quantile_data = census_data.groupby([f'county_{year}'])[f'fractionEducated_{year}', f'homeValuePercentChange_{year}'].quantile(2/3)
#     census_data = census_data.merge(quantile_data, on=f'county_{year}', suffixes= (None, '_quantile'))

#     education = census_data[f'fractionEducated_{year}'] > census_data[f'fractionEducated_{year}_quantile']
#     increasedHomeValue = census_data[f'medianHomeValue_{year}'] > census_data[f'medianHomeValue_{year - years_apart}']
#     homeValuePercentChange = census_data[f'homeValuePercentChange_{year}']
#     homePercentChangeQuantile = census_data[f'homeValuePercentChange_{year}'] > census_data[f'homeValuePercentChange_{year}_quantile']
#     ret = (education & increasedHomeValue & homePercentChangeQuantile & (homeValuePercentChange > 0) & census_data[f'isEligible_{year - years_apart}'])
#     assert ret is not None
#     return ret

def hasGentrified(census_data, years_apart, year):
    education = (census_data[f'fractionEducated_{year}'] >= census_data[f'fractionEducated_{year}'].quantile(0.5))
    educationPercentChange = (census_data[f'fractionEducated_{year}'] / census_data[f'fractionEducated_{year-years_apart}'] - 1)
    educationPercentChangeQuantile = (educationPercentChange >= educationPercentChange.quantile(0.5))
    increasedHomeValue = (census_data[f'adjustedHomeValue_{year}'] >= census_data[f'adjustedHomeValue_{year-years_apart}'])
    homeValuePercentChange = (census_data[f'adjustedHomeValue_{year}'] / census_data[f'adjustedHomeValue_{year-years_apart}'] - 1)
    homePercentChangeQuantile = (homeValuePercentChange >= homeValuePercentChange.quantile(0.5))
    # print(education.corr(homePercentChangeQuantile)) 
    ret = (educationPercentChangeQuantile & increasedHomeValue & homePercentChangeQuantile & (homeValuePercentChange > 0) & census_data[f'isEligible_{year - years_apart}'])
    assert ret is not None
    return ret

years_apart = 5

for year in range(2009+years_apart, 2019):
    census_data[f'hasGentrified_{year}_{years_apart}_apart'] = hasGentrified(census_data, years_apart, year)
    print(census_data[f'hasGentrified_{year}_{years_apart}_apart'].value_counts())
    #education = (census_data[f'fractionEducated_{year}'] >= census_data[f'fractionEducated_{year}'].quantile(2/3))
    #print(f"{year}:")
    #print(census_data[f'hasGentrified_{year}_{years_apart}_apart'].corr(education))
    #print(census_data[f'medianHouseIncome_{year}'].corr(education))
    #print("\n")
```
